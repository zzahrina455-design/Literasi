```python
class Barang:
    def __init__(self, id_barang, nama):
        self.id_barang = id_barang
        self.nama = nama

class Gudang:
    def __init__(self):
        self.koleksi_barang = []

    def tambah_barang(self, barang_baru):
        self.koleksi_barang.append(barang_baru)
        print(f"✅ {barang_baru.nama} berhasil masuk gudang.")

    def tampilkan_semua(self):
        print("\n--- DAFTAR BARANG DI GUDANG ---")
        for item in self.koleksi_barang:
            print(f"ID: {item.id_barang} | Nama: {item.nama}")

    def cari_by_id(self, id_cari):
        for item in self.koleksi_barang:
            if item.id_barang == id_cari:
                return item
        return None

    def filter_nama(self, kata_kunci):
        hasil = [item for item in self.koleksi_barang if kata_kunci.lower() in item.nama.lower()]
        return hasil

    def hitung_total_aset(self):
        total = sum(item.harga for item in self.koleksi_barang)
        return total

    def urutkan_nama(self):
        self.koleksi_barang.sort(key=lambda x: x.nama)
        print("Sorting selesai.")

    def tambah_dengan_validasi(self, barang_baru):
        if any(item.id_barang == barang_baru.id_barang for item in self.koleksi_barang):
            print("❌ Error: ID sudah terdaftar!")
        else:
            self.koleksi_barang.append(barang_baru)

    def hapus_barang(self, id_hapus):
        objek = self.cari_by_id(id_hapus)
        if objek:
            self.koleksi_barang.remove(objek)
            print(f"🗑️ Barang ID {id_hapus} telah dihapus.")
        else:
            print("❌ Barang tidak ditemukan.")

    def update_nama(self, id_target, nama_baru):
        objek = self.cari_by_id(id_target)
        if objek:
            objek.nama = nama_baru
            print("📝 Nama barang berhasil diperbarui.")

gudang_utama = Gudang()
gudang_utama.tambah_barang(Barang("A1", "Laptop ASUS"))
gudang_utama.tambah_barang(Barang("B2", "Mouse Logi"))
gudang_utama.tampilkan_semua()
gudang_utama.update_nama("B2", "Mouse Logitech Wireless")
```

```python
class Barang:
    def __init__(self, id_barang, nama, harga):
        self.id_barang = id_barang
        self.nama = nama
        self.harga = harga

    def __str__(self):
        return f"| {self.id_barang:8} | {self.nama:20} | Rp{self.harga:10,} |"

def tambah_barang(self, barang_baru):
    if any(item.id_barang == barang_baru.id_barang for item in self.koleksi_barang):
        return False, "ID sudah terdaftar!"
    self.koleksi_barang.append(barang_baru)
    return True, "Berhasil ditambahkan."

def tampilkan_semua(self):
    if not self.koleksi_barang:
        print("\n [ Gudang Kosong ]")
        return
    print("\n" + "="*50)
    print(f"| {'ID':8} | {'NAMA BARANG':20} | {'HARGA':12} |")
    print("-" * 50)
    for item in self.koleksi_barang:
        print(item)
    print("="*50)

def main():
    app = Gudang()
    while True:
        print("\n>>> SISTEM MANAJEMEN GUDANG V.1.0 <<<")
        print("1. Tambah Barang Baru")
        print("2. Lihat Semua Inventaris")
        print("3. Cari Barang (ID)")
        print("4. Hapus Barang")
        print("5. Keluar")

        pilihan = input("Pilih Menu (1-5): ")

        if pilihan == "1":
            id_b = input("Masukkan ID: ")
            nama = input("Masukkan Nama: ")
            try:
                harga = int(input("Masukkan Harga: ")) # Try-Except for crash prevention [cite: 11]
                sukses, pesan = app.tambah_barang(Barang(id_b, nama, harga))
                print(pesan)
            except ValueError:
                print("❌ Error: Harga harus berupa angka!")

import json
import os

class Barang:
    def to_dict(self):
        return {"id_barang": self.id_barang, "nama": self.nama, "harga": self.harga}

class Gudang:
    def __init__(self, file_db="database.json"):
        self.file_db = file_db
        self.koleksi_barang = self.muat_data()

    def simpan_data(self):
        try:
            with open(self.file_db, 'w') as f:
                data = [item.to_dict() for item in self.koleksi_barang]
                json.dump(data, f, indent=4)
        except Exception as e:
            print(f"Gagal menyimpan: {e}")

    def muat_data(self):
        if not os.path.exists(self.file_db):
            return []
        try:
            with open(self.file_db, 'r') as f:
                data_list = json.load(f)
                return [Barang(d['id_barang'], d['nama'], d['harga']) for d in data_list]
        except:
            return []

    def tambah_barang(self, barang_baru):
        self.simpan_data()

    def hapus_barang(self, id_hapus):
        self.simpan_data()

    def tambah_barang(self, barang_baru):

        self.simpan_data()

    def hapus_barang(self, id_hapus):

        self.simpan_data()
```
